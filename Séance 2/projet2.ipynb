```python
# import des modules usuels
import numpy as np
import pandas as pd

# options d'affichage
pd.set_option("display.min_rows", 16)
```

```python
# chargement et traitement des données
geo = pd.read_csv("correspondance-code-insee-code-postal.csv",
                   sep=';',
                   usecols=range(11),
                   index_col="Code INSEE")
geo = geo.sort_index()
geo[['Latitude', 'Longitude']] = geo["geo_point_2d"].str.extract("(.*), (.*)").astype(float)
geo.head()
```

**Projet maison n° 1**

- Ajoutez une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.
- Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.
- Ajoutez une fonction de conversion pour pouvoir utiliser la première fonction avec un GPS (degrés, minutes, secondes).

```python
geo['CP Ville'] = geo['Code Postal'] + ' ' + geo['Commune']
geo.head()
```

```python
# fonction recherche de ville
def ville(lat, long):
    geo['Distance'] = np.sqrt((lat - geo['Latitude'])**2 + (long - geo['Longitude'])**2)
    city = geo.loc[geo['Distance'] == geo.Distance.min(),'Commune']
    return city
```

```python
# conversion degrés, minutes, secondes => décimal
def dms2dec(deg, mn, sec):
    return deg + mn / 60 + sec / 3600
```

```python
# on applique la fonction à une coordonnée tirée au hasard
# init du random
np.random.seed(0)

a, b = 41.5, 51.1  # latitude min et max de la France métropolitaine
lat = (b - a) * np.random.random() + a
a, b = -5.1, 9.5  # longitude min et max de la France métropolitaine
long = (b - a) * np.random.random() + a

ville(lat, long)
```

```python
# à partir de coordonnées GPS précises
ville(dms2dec(48, 42, 52), dms2dec(2, 14, 45))
```

```python
ville(dms2dec(48, 55, 19), dms2dec(2, 15, 12))
```

Sur Windows shapely peut nécessiter d'installer la dll "geos_c.dll" dans le répertoire "Library/bin" de votre environnement Python.

```python
import json
geo_shape = geo['geo_shape'].apply(lambda x : json.loads(x))
df = pd.json_normalize(geo_shape)
geo['geo_type'] = geo_shape.apply(lambda x : x['type'])
geo['geo_coord'] = geo_shape.apply(lambda x : x['coordinates'])
geo.head()
```

```python
geo['geo_coord_len'] = geo['geo_coord'].apply(len)
```

```python
from shapely.geometry import Polygon
```

```python
A = geo.loc[(geo['geo_coord_len']==2)&(geo['geo_type']=='Polygon'), ['Commune','geo_coord']]
A['Polygon_Big'] = A['geo_coord'].apply(lambda x : Polygon(x[0]))
A['Polygon_Small'] = A['geo_coord'].apply(lambda x : Polygon(x[1]))
```

```python
A.head()
```

```python
def enclave(a,b):
    return a.contains(b)

for i in range(len(A['Polygon_Big'])):
    print(enclave(A['Polygon_Big'][i],A['Polygon_Small'][i]))
```
